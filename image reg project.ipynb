```python
import skimage.segmentation as seg
import skimage
```

```python
def image_show(image,nrows=1,ncol=1,cmap='gray',**kwargs):
    fig,ax=plt.subplots(nrows=nrows,ncols=ncol,figsize=(16,16))
    ax.imshow(image,cmap='gray',**kwargs)
    ax.axis('off')
    return fig,ax
```

```python
astronaut_seg=seg.slic(astronaut)
image_show(skimage.color.label2rgb(astronaut_seg,astronaut,kind='avg'))
```
