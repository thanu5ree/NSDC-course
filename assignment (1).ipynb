```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.flipkart.com/search?q=watches+girls&sid=r18%2Cf13&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=watches+girls%7CWrist+Watches&requestId=3a303e0a-d6d4-47b4-99db-61dd5dbe534b&as-searchtext=watches"
```

```python
url
```

```python
web=requests.get(url)
```

```python
allow=requests.get(url)
```

```python
allow
```

```python
soup=BeautifulSoup(allow.text)
```

```python
soup
```

```python
brand_name=soup.select('._2WkVRV')
```

```python
brand_name
```

```python
BRAND_NAME=[]
```

```python
for i in brand_name:
    BRAND_NAME.append(i.text)
```
